```python
chat = Chat()
```

```python
chat.ask(
    'What do these two lines of html do?:     <link rel="stylesheet" href="https://pyscript.net/latest/pyscript.css" />\n'
    '<script defer src="https://pyscript.net/latest/pyscript.js"></script>')
```

```python
with open('pyscript.css') as f:
    css = f.read()

chat.ask('what does this stylesheet do?: ' + css, model='gpt-3.5-turbo-16k')
```
